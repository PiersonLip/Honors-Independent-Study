```python
# Set the base path relative to your current script or notebook
base_path = Path().resolve().parent / 'data'

# Read the datasets
# oneline = pd.read_hdf(base_path / '1e+00_Zsun_population_max.h5', key='oneline')
history = pd.read_hdf(base_path / '1e+00_Zsun_population_max.h5', key='history')
# XB = pd.read_csv(base_path / 'XB.csv', low_memory=False)
# LMXB = pd.read_csv(base_path / 'LMXB.csv', low_memory=False)
# HMXB = pd.read_csv(base_path / 'HMXB.csv', low_memory=False)
# contact = pd.read_csv(base_path / 'contact.csv', low_memory=False)
# detached = pd.read_csv(base_path / 'detached.csv', low_memory=False)
```

```python
end_df = pd.DataFrame()
end_df = history[history['event']  == 'END']

BH = 'BH'
NS = 'NS'
WD = 'WD'
detached = 'detached'
RLO1 = 'RLO1'
RLO2 = 'RLO2'

S1_BH_NS = end_df['S1_state'].isin([BH, NS]) #creating bool frames where these conditions or met or not met
S2_BH_NS = end_df['S2_state'].isin([BH, NS])
S1_or_S2 = S1_BH_NS | S2_BH_NS
nor_S1_or_S2 = ~ (S1_BH_NS & S2_BH_NS)

S1_WD_Cond =~ end_df['S1_state'].isin([WD])
S2_WD_Cond =~ end_df['S2_state'].isin([WD]) 

State_Cond = end_df['state'].isin([detached, RLO1, RLO2]) 

XB = pd.DataFrame()
XB = end_df[S1_or_S2 & nor_S1_or_S2 & S1_WD_Cond & S2_WD_Cond & State_Cond]
XB.to_csv(base_path / 'XB.csv', index=False)

end_df.to_csv(base_path / 'endDF.csv', index=False)

HMXB = pd.DataFrame()
HMXB = XB.query('S2_mass > 10')
HMXB.to_csv(base_path / 'HMXB.csv', index=False)

LMXB = pd.DataFrame()
LMXB = XB.query('3 < S2_mass < 8')
LMXB.to_csv(base_path / 'LMXB.csv', index=False)

contact = pd.DataFrame()
contact = end_df[end_df['state'] == 'contact']
contact.to_csv(base_path / 'contact.csv', index=False)

detached = pd.DataFrame()
detached = end_df[end_df['state'] == 'detached']
detached.to_csv(base_path / 'detached.csv', index=False)

NS_state = detached[detached['S1_state'].isin(['NS'])]
VelaX1Dataset = NS_state.query('S2_mass > 10')
VelaX1Dataset.to_csv(base_path / 'velax1dataset.csv', index=False)

V404Dataset = XB.query('S2_mass < 5')
V404Dataset.to_csv(base_path / 'v404dataset.csv', index=False)

```
